Google Search Analysis 


```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
trends = TrendReq()
```

```python
trends.build_payload(kw_list=["Machine Learning"])
data = trends.interest_by_region()
data = data.sort_values(by = "Machine Learning", ascending = False)

"" Data regarding the keyword Machine Learning 
we try to see how much the keyword is searched in different regions. ""
```

```python
data  = data.head(10).reset_index()
```

```python
data
```

```python
#Plot 1

data["Country"] = data["geoName"]
# data.drop("geoName")

plot=sns.barplot(data = data, x = "Country", y = "Machine Learning")
plt.setp(plot.get_xticklabels(), rotation=45)
plt.show()
```

```python
data_1 = TrendReq(hl='en-US', tz=350)
data_1

data_1.build_payload(kw_list=['Machine Learning'])
data_1 = data_1.interest_over_time()

data_1 = data_1.sort_values(by = "date", ascending = False)

data_1
```

```python
#Plot 2

sns.lineplot(data = data_1, x = "date", y = "Machine Learning")
```

Remarks:

In this notebook the goal was to measure how the trends for the keyword Machine Learning varies based on region and time. 

A) Plot 1 shows that China is by far first the first country searching for Machine Learning.

B) In the Plot 2 it is clear that after 2022 there is a huge increase in Machine Learning search on Google.

C) This type of analysis can be used by Digital Marketers to optimize their keywords for SEO.
